**Rules:**
1. Colour used **sparingly** — one categorical variable, no rainbow
2. If showing all continents, either use accessible palette OR grey all + highlight one
3. `size_max` set when using bubble size
4. Log scale for GDP per capita
5. Insight title

```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
# Task 1 — Scatter: GDP vs life expectancy in Asia, comparing 2002 and 2007

# gapminder years closest to 2000 are 2002 and 2007
asia = df[(df['continent'] == 'Asia') & (df['year'].isin([2002, 2007]))].copy()
asia['year'] = asia['year'].astype(str)

colour_map = {'2002': '#AAAAAA', '2007': '#2196F3'}

fig = px.scatter(
    asia,
    x='gdpPercap', y='lifeExp',
    color='year',
    color_discrete_map=colour_map,
    hover_name='country',
    log_x=True,
    labels={'gdpPercap': 'GDP per Capita (log scale)', 'lifeExp': 'Life Expectancy', 'year': ''},
    title='Asian countries grew richer and lived longer by 2007 — the shift is consistent across the board'
)

# 2002 smaller and faded, 2007 bolder to emphasise the direction of change
fig.update_traces(selector=dict(name='2002'), marker=dict(size=8, opacity=0.5))
fig.update_traces(selector=dict(name='2007'), marker=dict(size=10, opacity=0.85))

# direct label instead of legend
for year, label, xpos, ypos in [('2002', '2002', 900, 44), ('2007', '2007', 900, 41)]:
    fig.add_annotation(
        x=xpos, y=ypos,
        text=f'<b>{label}</b>',
        showarrow=False,
        font=dict(color=colour_map[year], size=12, family='Arial')
    )

fig.update_traces(showlegend=False)

fig.update_layout(
    plot_bgcolor='white', paper_bgcolor='white',
    font=dict(family='Arial', size=13),
    yaxis=dict(gridcolor='#EEEEEE'),
    xaxis=dict(gridcolor='#EEEEEE'),
    margin=dict(l=60, r=60, t=55, b=40),
    height=500, width=900
)

fig.show()
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2 — Bubble chart: highlighting Europe's tight cluster at the top
import math

df_2007 = df[df['year'] == 2007].copy()

highlight = 'Europe'
colour_map = {c: ('#1B6CA8' if c == highlight else '#DDDDDD') for c in df_2007['continent'].unique()}

fig = px.scatter(
    df_2007,
    x='gdpPercap', y='lifeExp',
    size='pop',
    color='continent',
    color_discrete_map=colour_map,
    hover_name='country',
    log_x=True,
    size_max=55,
    custom_data=['country', 'pop'],
    labels={'gdpPercap': 'GDP per Capita (log scale)', 'lifeExp': 'Life Expectancy',
            'pop': 'Population', 'continent': ''},
    title='Europe stands out: high wealth, long lives — and remarkably little variation between countries'
)

# grey bubbles behind, highlighted on top with full opacity
fig.update_traces(
    showlegend=False,
    hovertemplate='<b>%{hovertext}</b><extra></extra>'
)
fig.update_traces(
    selector=dict(name=highlight),
    marker=dict(opacity=0.8, line=dict(width=0.5, color='white'))
)
fig.update_traces(
    selector=lambda t: t.name != highlight,
    marker=dict(opacity=0.4)
)

# annotation pointing to the European cluster
fig.add_annotation(
    x=math.log10(28000), y=77,
    text='Europe<br>high wealth,<br>long lives,<br>low spread',
    showarrow=True, arrowcolor='#1B6CA8',
    arrowhead=2, ax=80, ay=60,
    font=dict(color='#1B6CA8', size=11, family='Arial'),
    bgcolor='white', bordercolor='#1B6CA8', borderwidth=1, borderpad=4
)

fig.update_layout(
    plot_bgcolor='white', paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    yaxis=dict(gridcolor='#EEEEEE'),
    xaxis=dict(gridcolor='#EEEEEE'),
    margin=dict(l=60, r=60, t=55, b=40),
    height=550, width=1000
)

fig.show()
```
