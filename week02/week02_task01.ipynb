```python
import pandas as pd
import os
import plotly.express as px
```

```python
cwd = os.getcwd()
file_path = os.path.join(cwd,"..","data","world_happiness_2023.csv")
df = pd.read_csv(file_path)
df.columns = ['Country', 'Region', 'Happiness_Score', 'GDP', 'Social_Support',
              'Life_Expectancy', 'Freedom', 'Generosity', 'Corruption']
df.head()
```

```python
region_avg = (
    df.groupby('Region')['Happiness_Score']
    .mean()
    .reset_index()
    .sort_values('Happiness_Score')
)
region_avg['Happiness_Score'] = region_avg['Happiness_Score'].round(2)

fig1 = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    text='Happiness_Score',
    color='Happiness_Score',
    color_continuous_scale='Blues',
    title='Western Europe leads happiness by nearly 3.5 points over South Asia',
    labels={'Happiness_Score': 'Avg Happiness Score', 'Region': ''},
)

fig1.update_traces(textposition='outside')
fig1.update_layout(
    plot_bgcolor='white',
    xaxis=dict(range=[0, 8.5], showgrid=True, gridcolor='#eeeeee'),
    yaxis=dict(showgrid=False),
    coloraxis_showscale=False,
    height=420,
)

fig1.show()
```
